```python
import cv2
import mediapipe as mp
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

cap = cv2.VideoCapture('face_video.mp4')
with mp_face_detection.FaceDetection(
    model_selection=0, min_detection_confidence=0.5) as face_detection:
    while cap.isOpened():
        success, image = cap.read()
        if not success:
          break
            
        image.flags.writeable = False
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = face_detection.process(image)

        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        if results.detections:
            for detection in results.detections:
                mp_drawing.draw_detection(image, detection)
                
        cv2.imshow('MediaPipe Face Detection', cv2.resize(image, None, fx=0.5, fy=0.5))
    
        if cv2.waitKey(1) == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

#얼굴을 찾고, 찾은 얼굴에표시를 해주기 위한 변수 정의
mp_face_detection = mp.solutions.face_detection # 얼굴 검출을 위한 face_detection 모욜 사용
mp_drawing = mp.solutions.drawing_utils # 얼굴 특징을 그리기 위한 drawing_utils 모듈 사용

# 동영상 파일 열기
cap = cv2.VideoCapture('face_video.mp4')

# 별도로 close 해줄 필요 없이 자동으로 with문 탈출하고 나면 파일이 닫히기 위해 사용
# model_selection : face detection 객체를 만드는데 모델을 총 두가지 중 하나 골라서 선택할 수 있음, 0 or 1)
    # 0 : 카메라로부터 2미터 이내 근거리 / 1 : 5m 정도 이내의 얼굴
# min_detection_confidence : ~%이상 되면 얼굴로 인식할지를 정의(Threshold) , 0.5 -> 50퍼센트 정도 얼굴로 확신
with mp_face_detection.FaceDetection(
    model_selection=0, min_detection_confidence=0.7) as face_detection:
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            
            #if loading a video, use 'break' instead of 'continue'
            break
        #image.flasg.writable을 False였다가 어떠한 작업 후 True로 바꿔짐
        image.flags.writeable = False
        #convert color를 통해 BGR -> RGB로 변경 (mediapie모듈이 RGB로 처리)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #이미지로부터 얼굴을 검출하여 결과를 반환 
        results = face_detection.process(image)

        # 이미지로부투 얼굴을 검출 후 False -> True로 변경
        image.flags.writeable = True
        #convert color를 통해 RGB -> BGR로 변경
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # results에서 검출된 얼굴이 있으면 반복문을 통해검출된 사람 얼굴만큼 draw (사각형, 점 등)
        if results.detections:
            for detection in results.detections:
                # mp_drawing.draw_detection(image, detection)
                # print(detection) # detection 결과 보기
                # score : 신뢰도 (얼굴 검출 확률)
                # location data : bounding_box를 통해 얼굴 검출
                # 6개의 keypoint정보 확인 (오른쪽 / 왼쪽 눈 / 코 끝 / 입 중심 / 오른쪽 귀 / 왼쪽 귀
                # label_id -> 다음 frame detection 정보

                #특정 위치 가져오기
                keypoints = detection.location_data.relative_keypoints
                right_eye = keypoints[0]
                left_eye = keypoints[1]
                nose_tip = keypoints[2]

                #이미지로부터 세로, 가로 크기 가져옴
                h,w, _ = image.shape # height, width, channel
                # 튜플형태 이미지 내에서 실제 좌표(x,y) 구하기
                right_eye = (int(right_eye.x * w), int(right_eye.y * h)) 
                left_eye = (int(left_eye.x * w), int(left_eye.y * h))
                nose_tip = (int(nose_tip.x * w), int(nose_tip.y * h))

                #양 눈에 동그라미 그리기
                cv2.circle(image, right_eye, 50, (255,0,0), 10, cv2.LINE_AA) #파랑
                cv2.circle(image, left_eye, 50, (0,255,0), 10, cv2.LINE_AA) # 초록
                #양 눈에 동그라미 그리기
                cv2.circle(image, nose_tip, 75, (0,255,255), 10, cv2.LINE_AA) # 노랑
        #resize한 영상 실행
        cv2.imshow('MediaPipe Face Detection', cv2.resize(image, None, fx=0.5, fy=0.5))
    
        if cv2.waitKey(1) == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

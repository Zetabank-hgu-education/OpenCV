## 사다리꼴 이미지 펼치기

```python
import cv2
import numpy as np
img = cv2.imread('newspaper.jpg')

width, height = (640, 240) # 가로 640, 세로 240 크기로 결과물 출력

#원근을 적용하기 위해서는 input값이 필요 : 그림에서 4개의 점을 말함 -> input4개점을 output으로 바꾸는 작업이 필요
#4개의 지점을 찾고 이것들을 변화해서 직사각형 형태로 바꿔줌
src = np.array([[511, 352], [1008,345], [1122,584], [455,594]], dtype=np.float32) #input 4개 지점
#좌상, 우상, 우하, 좌하(시계방향으로 4지점 정의)
dst = np.array([[0, 0], [width,0], [width,height], [0,height]], dtype=np.float32) #output 4개 지점

#getperspectivetransform : src영역을 dst영역으로 바꾸기위한 변환행렬을 얻는것, 변환matrix를 얻는 것
matrix = cv2.getPerspectiveTransform(src, dst) #matrix 얻어옴 -> matrix값을 통해 변환
#warpperspective : 변환 행렬값을 적용해가지고 최종 결과 이미지를 얻어오는 것. (result)
result = cv2.warpPerspective(img, matrix, (width, height)) # matrix 대로 변환을 함

cv2.imshow('img', img)
cv2.imshow('result', result)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

### 마우스 이벤트 등록

```python
# 마우스 클릭, 움직임, 등의 이벤트 행동을 했을때 어떤한 처리를 등록하여 처리
```

```python
import cv2

def mouse_handler(event, x, y, flags, param): 
    if event == cv2.EVENT_LBUTTONDOWN: #마우스 왼쪽 버튼 누름
        print('왼쪽 버튼 Down')
        print(x,y)
    elif event == cv2.EVENT_LBUTTONUP: # 마우스 왼쪽 버튼 땜
        print('왼쪽 버튼 Up')
        print(x,y)
    elif event == cv2.EVENT_LBUTTONDBLCLK: # 마우스 왼쪽 버튼 더블 클릭
        print('왼쪽 버튼 Double Click')
    # elif event == cv2.EVENT_MOUSEMOVE: # 마우스 이동
    #     print('마우스 이동')
    elif event == cv2.EVENT_RBUTTONDOWN: # 오른쪽 버튼 클릭
        print('오른쪽 버튼 Down')
img = cv2.imread('poker.jpg')
cv2.namedWindow('img') # img란 이름의 윈도우 생성
cv2.setMouseCallback('img', mouse_handler) #이벤트를 처리하기위한 핸드러를 적용
cv2.imshow('img', img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
import cv2

def mouse_handler(event, x, y, flags, param): # x,y좌표가 마우스 이벤트를 발생한 좌표가 됨
    if event == cv2.EVENT_LBUTTONDOWN: #마우스 왼쪽 버튼 누름
        print('왼쪽 버튼 Down')
        print(x,y)
    elif event == cv2.EVENT_LBUTTONUP: # 마우스 왼쪽 버튼 땜
        print('왼쪽 버튼 Up')
        print(x,y)
    elif event == cv2.EVENT_LBUTTONDBLCLK: # 마우스 왼쪽 버튼 더블 클릭
        print('왼쪽 버튼 Double Click')
    elif event == cv2.EVENT_MOUSEMOVE: # 마우스 이동
        print('마우스 이동')
    elif event == cv2.EVENT_RBUTTONDOWN: # 마우스 이동
        print('오른쪽 버튼 Down')
img = cv2.imread('poker.jpg')
cv2.namedWindow('img') # img란 이름의 윈도우를 먼저 만들어 두는 것,여기에 마우스 이벤트를 처리하기위한 핸들러 적용
cv2.setMouseCallback('img', mouse_handler) # 이미지라는 이름들 가진 윈도우에 어떤 이벤트를 처리하기위한 핸드러를 적용
cv2.imshow('img', img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

### 프로젝트 코드

```python
import cv2
import numpy as np

point_list = [] 
src_img = cv2.imread('poker.jpg')

COLOR = (255, 0, 255) #핑크

def mouse_handler(event, x, y, flags, param): # x,y좌표가 마우스 이벤트를 발생한 좌표가 됨
    if event == cv2.EVENT_LBUTTONDOWN: #마우스 왼쪽 버튼 누름
        point_list.append((x,y)) 
    
    for point in point_list: 
        cv2.circle(src_img, point, 15, COLOR, cv2.FILLED)
    if len(point_list) == 4: #리스트안 4개 좌표가 다 찍혔을시
        show_result()# 결과 출력

    cv2.imshow('img', src_img) 

def show_result():
    width, height = (530, 710) # 가로 640, 세로 240 크기로 결과물 출력
    src = np.float32(point_list) #point_list 4개 지점을 가지고 float32형태 어레이를 만들어 주는 것
    dst = np.array([[0, 0], [width,0], [width,height], [0,height]], dtype=np.float32) #output 4개 지점
    matrix = cv2.getPerspectiveTransform(src, dst) #matrix 얻어옴 -> matrix값을 통해 변환
    result = cv2.warpPerspective(src_img, matrix, (width, height)) # matrix 대로 변환을 함
    cv2.imshow('result', result)

cv2.namedWindow('img') # img란 이름의 윈도우 생성
cv2.setMouseCallback('img', mouse_handler) # 이벤트를 처리하기위한 핸드러를 적용
cv2.imshow('img', src_img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
import cv2
import numpy as np

point_list = [] # 여기서 event 마우스 왼쪽 버튼 클릭시 포인트 리스트에 x,y를 넣는 작업할 예정
src_img = cv2.imread('poker.jpg')

COLOR = (255, 0, 255) #핑크

def mouse_handler(event, x, y, flags, param): # x,y좌표가 마우스 이벤트를 발생한 좌표가 됨
    if event == cv2.EVENT_LBUTTONDOWN: #마우스 왼쪽 버튼 누름
        point_list.append((x,y)) #튜플 형태, point_list에 현재 마우스를 누른 그 좌표가 들어가게됨
    #마우스로 클릭 했을 때 클릭한 지점을 그 이미지에다가 바로 표시하도록 하는 방법 , 리스트에 몇개가 있는지 모르니 반복문 사용
    for point in point_list: # 중심점은 위 (x,y)가 있으니 포인트 사용 / 리스트에 하나씩 가져와 xy좌표를 중심점으로 처리해줌 / 반지름은 5
        cv2.circle(src_img, point, 15, COLOR, cv2.FILLED)
    if len(point_list) == 4: #리스트안 4개 좌표가 다 찍혔을시
        show_result()# 결과 출력

    cv2.imshow('img', src_img) #이미지에다 지금 이미지를 다시한번 써주는걸로 함 : 이래야 마우스 킬릭마다 똑같은 이미지창에 변경된 이미지를 계속 덮을수있음

def show_result():
    width, height = (530, 710) # 가로 640, 세로 240 크기로 결과물 출력
    src = np.float32(point_list) #point_list 4개 지점을 가지고 float32형태 어레이를 만들어 주는 것
    dst = np.array([[0, 0], [width,0], [width,height], [0,height]], dtype=np.float32) #output 4개 지점
    matrix = cv2.getPerspectiveTransform(src, dst) #matrix 얻어옴 -> matrix값을 통해 변환
    result = cv2.warpPerspective(src_img, matrix, (width, height)) # matrix 대로 변환을 함
    cv2.imshow('result', result)

cv2.namedWindow('img') # img란 이름의 윈도우를 먼저 만들어 두는 것,여기에 마우스 이벤트를 처리하기위한 핸들러 적용
cv2.setMouseCallback('img', mouse_handler) # 이미지라는 이름들 가진 윈도우에 어떤 이벤트를 처리하기위한 핸드러를 적용
cv2.imshow('img', src_img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
